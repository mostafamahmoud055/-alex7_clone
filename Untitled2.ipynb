```python
from sklearn.datasets import load_iris,_california_housing,load_breast_cancer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,mean_squared_error
from sklearn.svm import LinearSVR,SVR,LinearSVC,SVC
import pandas as pd
```

```python
X,y=_california_housing.fetch_california_housing(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
pl=Pipeline(
    [('scaler',MinMaxScaler()),
     ('estimator',SVR(C=5))]
)
```

```python
pl.fit(X_train,y_train)
```

```python
pr=pl.predict(X_test)
```

```python
y_test
```

```python
pl.score(X_test,y_test)
```

```python
mean_squared_error(y_test,pr)
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
pl=Pipeline(
    [('scaler',MinMaxScaler()),
     ('estimator',LinearSVC())]
)
pl.fit(X_train,y_train)
pr=pl.predict(X_test)
print(classification_report(y_test,pr))
```

```python
pl=Pipeline(
    [('scaler',MinMaxScaler()),
     ('estimator',SVC(kernel='rbf'))]
)
pl.fit(X_train,y_train)
pr=pl.predict(X_test)
print(classification_report(y_test,pr))
# 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
```
